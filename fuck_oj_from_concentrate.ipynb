```python
#Import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from textwrap import wrap
```

```python
#All data are generated according to subjective estimates
oj_vit_iq = np.random.normal(loc=195,scale=10,size=50)
fresh_oj_iq = np.random.normal(loc=115,scale=20,size=50)
oj_iq = np.random.normal(loc=75,scale=15,size=50)
vit_iq = np.random.normal(loc=140,scale=13,size=50)
no_vit_iq = np.random.normal(loc=65,scale=8,size=50)
neither_iq = np.random.normal(loc=30,scale=5,size=50)
```

```python
#Assign labels to the data
data = np.hstack((oj_vit_iq,fresh_oj_iq,oj_iq,vit_iq,no_vit_iq,neither_iq))
labels = np.hstack((np.repeat("Drinks Fresh-Squeezed Orange Juice & Takes Vitamins",50),
                    np.repeat("Drinks Fresh-Squeezed Orange Juice",50),
                    np.repeat("Drinks Orange Juice From Concentrate",50),
                    np.repeat("Takes Vitamins",50),
                    np.repeat("Does Not Take Vitamins",50),
                    np.repeat("Drinks Orange Juice From Concentrate & Does Not Take Vitamins",50)))

labels = ['\n'.join(wrap(l, 13)) for l in labels]

#Create the dataframe
df = pd.DataFrame((data,labels)).T
df.columns = ["Data","Label"]
```

```python
#### Initialize figure settings and plot the data
sns.set(style="whitegrid")
fig,ax = plt.subplots()
fig.set_size_inches(9,7)

ax = sns.boxplot(x="Label", y="Data",data=df)
ax = sns.swarmplot(x="Label",y="Data",data = df,color="black",alpha=.3)

plt.xlabel("Measures of Orange Juice and Vitamin Consumption",fontsize=16,labelpad=15)
plt.ylabel("Intelligence Quotient (IQ)",fontsize=16,labelpad=15)
plt.title("Intelligence as a Function of Orange Juice and Vitamin Consumption",fontsize=17,y=1.02)
plt.show()
```
